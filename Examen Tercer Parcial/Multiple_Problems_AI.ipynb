# Exercise: Solving a Simple KnapSack Problem

## Problem Description  
The **KnapSack Problem** is a classic optimization problem. The goal is to select items to pack into a knapsack in such a way that the total value is maximized without exceeding the weight capacity of the knapsack.

---

## Problem Statement  
You are given the following items, each with a specific value and weight:

| Item | Value ($) | Weight (kg) |
|------|-----------|-------------|
| 1    | 60        | 10          |
| 2    | 100       | 20          |
| 3    | 120       | 30          |

The maximum capacity of the knapsack is **150 kg**. Your task is to determine which items to include in the knapsack to maximize the total value without exceeding the weight limit.

---

## Instance Characteristics  
- **Number of items:** 3  
- **Knapsack capacity:** 50 kg  
- **Items available:**
  - **Item 1:** Value = $60, Weight = 10 kg  
  - **Item 2:** Value = $100, Weight = 20 kg  
  - **Item 3:** Value = $120, Weight = 30 kg  

---

## Objective  
Maximize the total value of the items selected, subject to the weight constraint. Mathematically, this can be formulated as:

La siguiente es una formulación de un problema de optimización lineal:

Maximizar:
$ 60x_1 + 100x_2 + 120x_3$ 

Sujeto a la restricción:
$ 10x_1 + 20x_2 + 30x_3 <= 50$ 

Donde:
- $x_i$ es una variable binaria (1 si el artículo i está incluido, 0 en caso contrario).


---

## Tasks  
1. **Problem Formulation:**  
   - Identify which items can be included without exceeding the weight limit.  
   - Write the mathematical model as described above.

2. **Solve the Problem:**  
Use a simple approach, such as:
- **Genetic Algorithm (GA)**  
- **Ant Colony Optimization (ACO)**  
- **Particle Swarm Optimization (PSO)**  
- **Simulated Annealing (SA)**  
- **Tabu Search**  
- **Greedy Randomized Adaptive Search Procedure (GRASP)**solution.

3. **Output:**  
   - List the selected items, their total value, and their total weight.
   - Graph the best solution obtained on each iteration




```python
# Importar bibliotecas necesarias
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Cargar el dataset
soil_crops = pd.read_csv('soil_crops.csv')

# Exploración de datos
print("Primeros registros del dataset:")
print(soil_crops.head())
print("\nEstadísticas descriptivas:")
print(soil_crops.describe())

# Visualizaciones
sns.pairplot(soil_crops, hue='Crop')
plt.title("Relación entre características y tipo de cultivo")
plt.show()

# Preparación de datos
# Convertir la columna SoilType a valores numéricos
encoder = OneHotEncoder(sparse_output=False)  # Cambiado de sparse a sparse_output
soil_type_encoded = encoder.fit_transform(soil_crops[['SoilType']])

# Crear un DataFrame con las características codificadas
soil_type_df = pd.DataFrame(soil_type_encoded, columns=encoder.get_feature_names_out(['SoilType']))

# Combinar las características codificadas con el resto del DataFrame
X = pd.concat([soil_crops[['pH', 'Moisture', 'Nitrogen']], soil_type_df], axis=1)
y = soil_crops['Crop']

# Escalado de características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# División de datos
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Modelo KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Evaluación del modelo
y_pred = knn.predict(X_test)
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred))
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))
```

```python
# Importar bibliotecas necesarias
from sklearn.preprocessing import StandardScaler
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_score
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
customer_data = pd.read_csv('customer_data.csv')

# Exploración de datos
print(customer_data.describe())
sns.pairplot(customer_data)
plt.show()

# Preparación de datos
scaler = StandardScaler()
customer_data_scaled = scaler.fit_transform(customer_data)

# Determinación del número óptimo de clusters (Elbow method adaptado para K-Medoids)
inertia = []
for k in range(1, 10):
    kmedoids = KMedoids(n_clusters=k, random_state=42, method="pam")
    kmedoids.fit(customer_data_scaled)
    inertia.append(kmedoids.inertia_)

plt.plot(range(1, 10), inertia, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.title('Método del codo (K-Medoids)')
plt.show()

# Clustering con K-Medoids
kmedoids = KMedoids(n_clusters=4, random_state=42, method="pam")
customer_data['Cluster'] = kmedoids.fit_predict(customer_data_scaled)

# Visualización 2D
fig = px.scatter(customer_data, x='MonthlyIncome', y='AverageSpending', color='Cluster', title="Clusters de Clientes (K-Medoids)")
fig.show()

# Evaluación
print(f"Silhouette Score: {silhouette_score(customer_data_scaled, customer_data['Cluster'])}")
```

```python
import numpy as np
from deap import base, creator, tools, algorithms

# Datos del problema
values = [60, 100, 120]
weights = [10, 20, 30]
max_weight = 50

# Configuración del algoritmo genético
# 1. Definir una función de evaluación
def eval_knapsack(individual):
    total_value = sum(individual[i] * values[i] for i in range(len(individual)))
    total_weight = sum(individual[i] * weights[i] for i in range(len(individual)))
    if total_weight > max_weight:
        return 0,
    return total_value,

# 2. Crear la población y configurar el problema
creator.create("FitnessMax", base.Fitness, weights=(1.0,))  # Maximizar valor
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_bool", np.random.randint, 0, 2)  # Cada artículo puede incluirse (1) o no (0)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=len(values))
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("evaluate", eval_knapsack)
toolbox.register("mate", tools.cxTwoPoint)  # Cruce de dos puntos
toolbox.register("mutate", tools.mutFlipBit, indpb=0.1)  # Mutación por cambio de bit
toolbox.register("select", tools.selTournament, tournsize=3)

# 3. Ejecutar el algoritmo genético
population = toolbox.population(n=50)  # Tamaño de la población
ngen = 50  # Número de generaciones
cxpb, mutpb = 0.5, 0.2  # Probabilidad de cruce y mutación

result = algorithms.eaSimple(population, toolbox, cxpb, mutpb, ngen, verbose=False)

# Obtener el mejor individuo
best_individual = tools.selBest(population, k=1)[0]
selected_items = [i for i, bit in enumerate(best_individual) if bit == 1]
total_value = sum(values[i] for i in selected_items)
total_weight = sum(weights[i] for i in selected_items)

# Mostrar resultados
print(f"Mejor combinación de artículos: {selected_items}")
print(f"Valor total: {total_value}")
print(f"Peso total: {total_weight}")
```

Artículo 2: Valor = $100, Peso = 20 kg
Artículo 3: Valor = $120, Peso = 30 kg
El algoritmo genético encontró la mejor combinación de artículos, seleccionando los artículos 2 y 3, con un valor total de $220 y un peso total de 50 kg (el límite de peso).

# Análisis
Función objetivo
La función objetivo se definió para maximizar el valor total de los artículos seleccionados mientras se respeta la restricción de peso. Los artículos que exceden el límite de peso reciben un puntaje de 0, lo que ayuda al algoritmo a evitar combinaciones inviables.

# Configuración del algoritmo genético

Población inicial: Se generaron 50 individuos al azar.
Cruce: Se utilizó un cruce de dos puntos para combinar genes de los padres y generar nuevos individuos.
Mutación: Se aplicó una mutación de cambio de bit con una probabilidad de 10%, introduciendo diversidad en la población.
Selección: El torneo de selección permitió elegir individuos con mejor desempeño para la siguiente generación.
Resultados
El modelo identificó una combinación óptima de artículos que maximiza el valor dentro de las restricciones. Esto demuestra la capacidad de los algoritmos genéticos para abordar problemas de optimización combinatoria como el Knapsack Problem.

# Limitaciones
Convergencia local
Aunque el resultado obtenido es óptimo para el conjunto de datos actual, los algoritmos genéticos pueden quedar atrapados en máximos locales dependiendo de la configuración de parámetros.
