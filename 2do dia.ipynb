## Tarea
- Agarrar almenos 1 base de datos y generar una grafica de cada una de las que vimos



```python
import pandas as pd
import arff
import matplotlib.pyplot as plt

# Cargar el archivo ARFF (Pendigits)
with open('dataset_32_pendigits.arff') as f:
    dataset_pendigits = arff.load(f)

# Convertir los datos a DataFrame de pandas
df_pendigits = pd.DataFrame(dataset_pendigits['data'], columns=[attr[0] for attr in dataset_pendigits['attributes']])

# Ver los primeros registros del dataset
df_pendigits.head()
```

```python
# Grafica de barras
df_pendigits['class'].value_counts().plot(kind='bar', color='skyblue')

plt.title('Frecuencia de cada dígito en el dataset Pendigits')
plt.xlabel('Dígito')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Grafica de pastel
df_pendigits['class'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#c2f0c2','#ff6666','#66ff66','#ffcc66'])

plt.title('Distribución de los dígitos en el dataset Pendigits')
plt.ylabel('')
plt.show()
```

```python
# Grafica de dispersion
plt.scatter(df_pendigits['input1'], df_pendigits['input2'], alpha=0.5, color='purple')

plt.title('Dispersión de Input1 vs Input2')
plt.xlabel('Input1')
plt.ylabel('Input2')
plt.show()
```

```python
# grafica de histograma
df_pendigits['input1'].plot(kind='hist', bins=20, color='lightgreen')

plt.title('Distribución de Input1')
plt.xlabel('Input1')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Grafica de lineas
df_pendigits.groupby('class')[['input1', 'input2', 'input3', 'input4', 'input5', 'input6', 'input7', 'input8', 'input9', 'input10', 'input11', 'input12', 'input13', 'input14', 'input15', 'input16']].mean().plot(kind='line', marker='o')

plt.title('Promedio de inputs por clase (dígito)')
plt.xlabel('Dígito')
plt.ylabel('Promedio de Inputs')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
# Grafico box plot
df_pendigits.boxplot(column=['input1', 'input2', 'input3', 'input4'], by='class', grid=False, figsize=(10, 6))

plt.title('Distribución de Inputs por clase (dígito)')
plt.suptitle('')
plt.xlabel('Dígito')
plt.ylabel('Valores de Input')
plt.show()
```
