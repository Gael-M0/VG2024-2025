# Numeros aleatorios
Son numeros que deben cunplir ciertas caracteristicas y tener la misma probabilidad de ser elegidos y que la eleccion no dependa de algun otro
En otras palabras son numeros que son generados a partir de una distribucion uniforme

## Propiedades de numeros aleatorios
Su generacion se basa en el uso de mecanismos fisicos
tienen el inconveniente de ser generados lentamente
Son muy costosos de reproducir

## Caracteristicas numeros aleatorios
- Son numeros equiprobaables
- Son independientes
- son improbables
