## Analisis exploratorio de los datos / Analisis descriptivo

### Tipos de datos
- Cualitativos. Se describen por caracteristicas generales. Categoricos
- Cuantitativos. Representa algo que se puede contar

### Tipos de Variables / tipos de datos cuantitativos
- Discreto: Numero finito de valores
- Continuo: Numero infinito de valores en un intervalo
- Nominales: Los valores que permiten clasificar a una poblacion en categorias excluyentes
- Ordinales: Valores que pueden ser ordenados

```python
# Tarea 1 ejercicio
import pandas as pd
import arff

# Función para clasificar las variables
def clasificar_variable(col):
    if pd.api.types.is_numeric_dtype(col):
        if col.nunique() < 10:
            return "Discreto"
        else:
            return "Continuo"
    elif isinstance(col.dtype, pd.CategoricalDtype) or pd.api.types.is_object_dtype(col):
        if col.nunique() < 10:
            return "Nominal"
        else:
            return "Ordinal"
    return "Desconocido"

# Cargar el archivo ARFF
with open('dataset_32_pendigits.arff') as f:
    dataset_pendigits = arff.load(f)

# Convertir los datos a un DataFrame de pandas
df_pendigits = pd.DataFrame(dataset_pendigits['data'], columns=[attr[0] for attr in dataset_pendigits['attributes']])

# Mostrar los primeros registros del dataset
print("Primeros registros del dataset Pendigits:")
print(df_pendigits.head())

# Clasificar las variables según el tipo
clasificacion_pendigits = {col: clasificar_variable(df_pendigits[col]) for col in df_pendigits.columns}

# Mostrar la clasificación de tipos de variables
print("\nClasificación de tipos de variables en el dataset Pendigits:")
for col, tipo in clasificacion_pendigits.items():
    print(f"{col}: {tipo}")
```
